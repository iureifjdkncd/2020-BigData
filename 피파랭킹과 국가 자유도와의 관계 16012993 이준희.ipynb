#### 피파랭킹은 데이터셋의 변수 중에서 pf_association_sport와 가장 직적접으로 연결이 있습니다. 이를 바탕으로 과연 자유도가 높은 국가일수록 스포츠에 관한 자유도가 높고 이를 바탕으로 피파랭킹을 대입하여 추론을 했습니다.

##### 또한 이 프로젝트는 2017 피파랭킹 표와 함께 보면 더욱 이해가 잘 됩니다.

#### 이 변수들은 선진국으로써 갖춰야 할 가장 기본적인 요소들이라 판단했기에 추출했습니다.

#### 다음 단계를 진행하기 위해서 데이터셋의 고유 특성으로 처음에 숫자로 열이 채워진 변수들도 전부 object라고 나와 있어서 이를 숫자에 맞게 int() 혹은 float()로 변환했습니다.

```python
import numpy as np
```

```python
HighFreedom = fr[(fr["hf_score"]>=8) & (fr["pf_association_sport"]>=7.5) & (fr["year"]>=2017)] 
```

#### 기본적으로 자유도가 최소 8이상인 국가를 자유도가 높은 국가들이라고 할 수 있습니다. 자유도가 8~10의 점수에 해당되면 이제 그 안에서의 점수차이는 거의 의미가 없고 여기에 해당되는 국가들이 선진국이라는 것에 의미를 두는 것이 중요합니다.

#### 또한, 스포츠 자유도를 7.5이상의 준수한 수준으로 설정해서 전세계적으로 상대적으로 스포츠 관련 국제대회의 성적이나 스포츠산업 투자가 활발한 국가들로 범위를 압축했습니다.

#### 마지막으로 러시아 월드컵 출전 하기 전 해인 2017년의 데이터로 범위를 축소해서 실제 2017년 당시 피파랭킹 순위로 선택한 변수들을 함께 사용해 어떠한 국가들이 단지 축구를 잘하는 이유를 넘어서 추가적으로 어떠한 이유로 랭킹이 높았는지 살펴볼 수 있습니다.

#### 이 모든 조건으로 맞춘 데이터셋의 이름을 최종으로 HighFreedom이라고 설정했습니다.

#### 데이터셋에서 과연 어떠한 지역들이 주로 위의 조건들을 충족하는지 알아보기 위해서 ["region"]을 별도로 추출한 결과, 놀랍게도 유럽국가들의 빈도가 압도적으로 많이 나왔습니다.

#### 그 외에 몇몇 다른 대륙의 국가들이 출현했지만, 이것은 어디까지나 스포츠의 자유도가 높을수록 상대적으로 피파랭킹이 높다는 것이지, 모든 국가가 축구를 잘한다는 의미는 아닙니다.

#### 실제로 축구강국들이 대부분 유럽에 포진되어 있고 세계 최고 수준의 리그들이 영국,스페인,이탈리아에 있음은 물론이고 이 리그들로 타 국가의 선수들이 상당히 많이 이적해서 본인들의 커리어를 쌓고 인정을 받습니다.

#### hf_score을 여러 변수와 비교해서 groupby()를 통해서 숫자들의 관계를 나타낸 결과, 자유도는 상대적으로 스포츠,종교,법적보호,자산증대,거래비중에 있어서 거의 양의 관계를 보입니다.

#### 물론, 각 국가별로 정치,경제,안보의 측면에서 제도들이 확실히 다르지만, 자유도 점수로 이렇게 군집을 해서 분류를 했을 때 어느정도 일관된 흐름이 있습니다.

#### 군사 및 노동시간 같은 경우 흐름에 일관성을 띄기 보다 각 국가의 제도에 충실한 것으로 보입니다. 사실 충분히 이러한 결과가 나올 수 있는 이유는 예시로 설명할 수 있습니다. 

#### 우리나라의 경우, 세계적으로 군사력은 상위권에 있고, 평균 노동 및 근로시간 또한 많은 것으로 상위권입니다. 하지만 여러 방면에서 살기 좋은 자유 민주주의 국가이고 정치 및 경제적 자유도가 높습니다. 그러나 스포츠 관련해서 국제적인 성적은 조금 아쉬운 점이 있고 양궁,태권도,쇼트트랙 같은 종목에서도 이제 웬만하면 독보적인 존재감을 나타내지 않습니다. 나아가 축구에서는 특히 국민들의 열광이 힘찬 반면, 아시아에서는 확실히 강호의 모습을 보이지만,국가대표 A매치 성적은 유럽 혹은 남미의 강팀을 만나면 대부분 패배의 전적이 있습니다.

#### pivot으로 이번에 ef_money_growth, 즉 경제성장과 어느정도 연관이 있는 변수를 설정해서 스포츠와 국가들을 기준으로 두고 money growth의 가치를 나타냈을 때에도 역시 웬만하면 피파랭킹에서 어느정도 높은 위치에 해당하는 국가들이 경제력 역시 상위권에 주로 포진되어 있음을 알 수 있습니다.

#### 경제력과 스포츠 자유도가 높은 점을 보아 이 국가들은 스포츠 산업에 많은 투자를 하고 있고, 특히 유럽 국가들은 전통적으로 축구가 하나의 문화이기 때문에 더욱 자국리그와 국가대표팀에 투자를 많이 한다는 것을 알 수 있습니다.

```python
small = HighFreedom[["countries","pf_association_sport","ef_money_growth"]]
```

```python
small.pivot(index="pf_association_sport",columns="countries",values ="ef_money_growth").head()
```

#### 끝으로 이 프로젝트를 통해서 기본적으로 선진국들이 보다 피파랭킹에서 높은 순위를 자랑했던 것이 자유도와 상대적으로 많이 연관이 되어 있었음을 알 수 있었습니다

#### 물론 세상에 완벽한 데이터나 완벽한 설명이 없듯이 간혹 변수가 존재하지만, 뭐든지 절대적으로 흑백논리로 배분하는 것보다 상대적인 특성을 가진 결과들이 더욱 의미가 있고 과거,현재,미래와의 관계 또한 유기적으로 연결이 되도록 돕는 것 같습니다.
