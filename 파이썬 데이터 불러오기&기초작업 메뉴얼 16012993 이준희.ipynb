```python
import pandas as pd #데이터 프레임 및 csv 파일을 불러오고 작업을 하기 위해서 pandas를 입력해야 한다. 
```

```python
df = pd.read_csv("Downloads/BostonHousing.csv") #csv 파일을 불러온다.
```

```python
df.head() #파일의 상단에 위치한 변수들의 값을 나타낸다. 하단의 변수들은 .tail()을 입력하면 된다. 
```

```python
#df = df.replace("-", "0") #만약 결측치가 있으면 이런식으로 0으로 통일시키거나 아예 ""으로 공란으로 둘 수 있다. 
```

```python
select = df["MEDV"] #변수를 선택할 때는 []를 사용해서 안에 원하는 변수를 입력한다. 
```

```python
df = df[["MEDV","CRIM","CHAS","RM","TAX","NOX"]] # 다수의 변수를 선택하고 싶으면 [[]]를 사용해야 한다. 
```

```python
del df["RM"] #변수를 삭제하고 싶으면 del 함수를 사용하면 된다. 
```

```python
df.head() # 이렇게 해서 특정 변수들을 선택하고 삭제를 한 데이터 프레임으로 축소할 수 있다. 
```

```python
df[["MEDV","CRIM"]].groupby("MEDV").mean().head() #groupby를 통해서 기준이 되는 변수를 설정해서 기준변수로 다른 변수들을 비교분석 할 수 있다.
```

```python
df.groupby("CRIM").mean().head() #예시로, CRIM 기준으로 나머지 변수들이 어떠한 관계를 가지는지 볼 수 있다. 
```

```python
df.groupby(["MEDV","CRIM"])["TAX"].mean().head() # 두 변수를 함께 활용해서 groupby()를 할 수 있다. 이후에 원하는 변수를 []안에 입력해서 groupby()기준으로 해당 값들을 볼 수 있다.
```

```python
df.sort_values(by="MEDV").head(10) #sort()를 통해서 특정 변수를 순서대로 맞추면 전체 변수들 역시 특정 변수의 순서에 영향을 받아 전체 행의 순서가 맞춰진다.
```

```python
df.sort_values(by=["MEDV"],ascending=True).head(10) #ascending은 작은 수 부터 순서를 정하는 것이고 이것은 기본값이기 때문에 큰 수부터의 역순을 원하면 ascending = False라고 입력할 수 있다.
```

```python
df.sort_values(by=["MEDV"],ascending=False).head(10)
```

```python
#pivot의 경우 여러 데이터 중에서 자신이 원하는 데이터만을 가지고 원하는 행과 열에 데이터를 배치하는 작업이다.
```

```python
df.pivot(index="MEDV",columns="CRIM",values="TAX") #MEDV와 CRIM 기준으로 TAX의 가치를 보여준다.
```
